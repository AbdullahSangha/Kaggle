# Exploratory Data Analysis 

## Top 20 Countries with consecutive medals

```python
plt.figure(figsize=(24,10))
ax_1 = sns.barplot(x = 'team', y = 'total', data=data.head(20), palette='rainbow')

#displays bar labels
for i in ax_1.containers:
    ax_1.bar_label(i, size=16)

plt.xticks(fontsize=13, color='midnightblue', rotation=45)
plt.yticks(fontsize=13, color='midnightblue')
plt.xlabel('Country', size = 18, color='midnightblue')
plt.ylabel('Total Medals', size = 17 , color='midnightblue')
plt.title('Top 20 Countries W.R.T consecutive medals', size = 22 , color='midnightblue')

plt.show()
```

## Top 20 Countries W.R.T Gold Medals

```python
plt.figure(figsize=(24,10))
ax_1 = sns.barplot(x = 'team', y = 'gold', data=data.head(20), palette=my_colors_1)

#displays bar labels
for i in ax_1.containers:
    ax_1.bar_label(i, size=16)

plt.xticks(fontsize=13, color='midnightblue', rotation=45)
plt.yticks(fontsize=13, color='midnightblue')
plt.xlabel('Country', size = 18, color='midnightblue')
plt.ylabel('Total Gold Medals', size = 17 , color='midnightblue')
plt.title('Top 20 Countries W.R.T Gold Medals', size = 22 , color='midnightblue')

plt.show()
```

## Top 20 Countries W.R.T Silver Medals

```python
plt.figure(figsize=(24,10))
ax_1 = sns.barplot(x = 'team', y = 'silver', data=data.head(20), palette=my_colors)

#displays bar labels
for i in ax_1.containers:
    ax_1.bar_label(i, size=16)

plt.xticks(fontsize=13, color='midnightblue', rotation=45)
plt.yticks(fontsize=13, color='midnightblue')
plt.xlabel('Country', size = 18, color='midnightblue')
plt.ylabel('Total Silver Medals', size = 17 , color='midnightblue')
plt.title('Top 20 Countries W.R.T Silver Medals', size = 22 , color='midnightblue')

plt.show()
```

## Scatter cross Gold and Total

```python
fig = px.scatter(data, x="total", y="gold", color="gold", size_max=35,
                 size='gold', template = 'plotly_white', hover_data=['team', 'gold', 'silver', 'bronze'],
                 title='Scatter Plot Cross Gold and Total')
fig.show()
```

## Scatter Geo of Total Medals

```python
fig = px.scatter_geo(data, locationmode = "country names", locations="team", color="total",
                     hover_data=['team', 'gold', 'silver', 'bronze'], size="total",
                     projection="natural earth", size_max = 25, template = 'plotly_dark')
fig.show()
```
