## Top 20 Countries in terms of Best Quality of Life

```python
df_20 = data.head(20)
fig = px.bar(df_20, x='Country', y='TotalQuality of life(100%)', 
             template = 'plotly_white', title='Top 20 Countries in terms of Best Quality of Life', 
             color='TotalQuality of life(100%)', height=800, 
             hover_data= ['Rights(20%)', 'Health(15%)', 'Stability(15%)'])
fig.update_xaxes(tickangle=45, rangeselector_font_size=8)
fig.show()
```

## Top 20 Countries in terms of Best Quality of Life

```python
plt.figure(figsize=(24,10))
ax_1 = sns.barplot(x=data['Country'].head(20), y=data['TotalQuality of life(100%)'], data=data, palette=my_colors_1)

plt.xticks(fontsize=13, color='midnightblue', rotation=45)
plt.yticks(fontsize=13, color='midnightblue')
plt.xlabel('Country', size = 18, color='midnightblue')
plt.ylabel('Quality of life', size = 17 , color='midnightblue')
plt.title('Top 20 Countries in terms of Best Quality of Life', size = 22 , color='midnightblue')
plt.show()
```

## Top 20 Countries in terms of Worst Quality of Life

```python
dft_20 = data.tail(20)
fig = px.bar(dft_20, x='Country', y='TotalQuality of life(100%)', 
             template = 'plotly_white', title='Top 20 Countries in terms of Worst Quality of Life', 
             color='TotalQuality of life(100%)', height=800, 
             hover_data= ['Rights(20%)', 'Health(15%)', 'Stability(15%)'])
fig.update_xaxes(tickangle=45, rangeselector_font_size=8)
fig.show()
```

## Top 20 Countries in terms of Worst Quality of Life

```python
plt.figure(figsize=(24,10))
ax_1 = sns.barplot(x=data['Country'].tail(20), y=data['TotalQuality of life(100%)'], data=data, palette=my_colors_1)

plt.xticks(fontsize=13, color='midnightblue', rotation=45)
plt.yticks(fontsize=13, color='midnightblue')
plt.xlabel('Country', size = 18, color='midnightblue')
plt.ylabel('Quality of Life', size = 17 , color='midnightblue')
plt.title('Top 20 Countries in terms of Worst Quality of Life', size = 22 , color='midnightblue')
plt.show()
```

## Scatter Plot of cross Stability & Rights
